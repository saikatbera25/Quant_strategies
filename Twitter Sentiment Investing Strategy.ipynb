# Twitter Sentiment Investing Strategy  

# 1. Load Twitter Sentiment Data  

Load the twitter sentiment dataset, set the index, calculate the engagement ratio and filter out stocks with no significant twitter activity.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import datetime as dt
import yfinance as yf
import os
plt.style.use('ggplot')

```

```python
# Offline data stored as a CSV file, so specifying path to that file

data_folder = 'C:/Users/saika/Downloads/Python_files/Quant_Strategies/Twitter_Sentiment_Data_Files'
```

```python
sentiment_df = pd.read_csv(os.path.join(data_folder, 'sentiment_data.csv'))

sentiment_df
```

```python
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])

sentiment_df
```

```python
sentiment_df = sentiment_df.set_index(['date', 'symbol'])

sentiment_df
```

```python
# Calculating/Defining and attribute to gauge the engagement

sentiment_df['engagement_ratio'] = sentiment_df['twitterComments']/sentiment_df['twitterLikes']

sentiment_df
```

```python
sentiment_df = sentiment_df[(sentiment_df['twitterLikes']>20)&(sentiment_df['twitterComments']>10)]

sentiment_df
```

# 2. Aggregate Monthly and calculate average sentiment for the month  

Aggregate on a monthly level and calculate average monthly metric, for the one we choose  

We will select to invest in the top five stocks every month for the next month

```python
sentiment_df.reset_index('symbol')
```

```python
sentiment_df.reset_index('symbol').groupby([pd.Grouper(freq='M'), 'symbol'])

sentiment_df
```

```python
(sentiment_df.reset_index('symbol').groupby([pd.Grouper(freq='M'), 'symbol'])
[['engagement_ratio']].mean())
```

```python
aggregated_df = (sentiment_df.reset_index('symbol').groupby([pd.Grouper(freq='M'), 'symbol'])
                [['engagement_ratio']].mean())

aggregated_df
```

```python
aggregated_df['rank'] = (aggregated_df.groupby(level=0)['engagement_ratio']
                          .transform(lambda x: x.rank(ascending=False)))

aggregated_df
```

# 3. Select the top 5 stocks based on their cross-sectional ranking for each month  

Select top 5 stocks by rank for each month and fix the date to start at begining of next month

```python
aggregated_df[aggregated_df['rank']<6]
```

```python
filtered_df = aggregated_df[aggregated_df['rank']<6].copy()

filtered_df
```

```python
filtered_df = filtered_df.reset_index(level=1)

filtered_df
```

```python
filtered_df.index = filtered_df.index+pd.DateOffset(1)

filtered_df
```

```python
filtered_df = filtered_df.reset_index().set_index(['date', 'symbol'])

filtered_df.head(20)
```

# 4. Extract the stocks to form portfolios with at the start of each new month  

Create a dictionary containing start of month and corresponding selected stocks

```python
dates = filtered_df.index.get_level_values('date').unique().tolist()

fixed_dates = {}

for d in dates:
    
    fixed_dates[d.strftime('%Y-%m-%d')] = filtered_df.xs(d, level=0).index.tolist()
    
fixed_dates
```

# 5. Download fresh stock prices for only selected/shortlisted stocks

```python
stocks_list = sentiment_df.index.get_level_values('symbol').unique().tolist()

prices_df = yf.download(tickers=stocks_list,
                        start='2021-01-01',
                        end='2023-03-01')
```

```python
prices_df
```

# 6. Calculate Portfolio Return with monthly rebalancing

```python
returns_df = np.log(prices_df['Adj Close']).diff().dropna()

returns_df
```

The above issue is happening because the ['ATVI'] stock has been delisted and somehow because of that every row in the dataframe gets a NaN value  

To resolve that, remove the ['ATVI'] stock from the stocks_list and then get the prices data as shown below

```python
stocks_list = sentiment_df.index.get_level_values('symbol').unique().tolist()

stocks_list.remove('ATVI')

prices_df = yf.download(tickers=stocks_list,
                        start='2021-01-01',
                        end='2023-03-01')
```

```python
returns_df = np.log(prices_df['Adj Close']).diff().dropna()

returns_df
```

```python
portfolio_df = pd.DataFrame()

for start_date in fixed_dates.keys():
    
    end_date = (pd.to_datetime(start_date)+pd.offsets.MonthEnd()).strftime('%Y-%m-%d')
    
    cols = fixed_dates[start_date]
    
    temp_df = returns_df[start_date:end_date][cols].mean(axis=1).to_frame('portfolio_return')
    
    portfolio_df = pd.concat([portfolio_df, temp_df], axis=0)
    
portfolio_df
```

# 7. Download NASDAQ/QQQ prices and calculate returns to compare to our strategy

```python
qqq_df = yf.download(tickers='QQQ',
                     start='2021-01-01',
                     end='2023-03-01')

qqq_ret = np.log(qqq_df['Adj Close']).diff().to_frame('nasdaq_return')

portfolio_df = portfolio_df.merge(qqq_ret,
                                  left_index=True,
                                  right_index=True)

portfolio_df
```

Now to visualize these two returns against each other

```python
portfolios_cumulative_return = np.exp(np.log1p(portfolio_df).cumsum()).sub(1)

portfolios_cumulative_return.plot(figsize=(16,6))

plt.title('Twitter Engagement Ratio Strategy Over Time')

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1))

plt.ylabel('Return')

plt.show()
```
